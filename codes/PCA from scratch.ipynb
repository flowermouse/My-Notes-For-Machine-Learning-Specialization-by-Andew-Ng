```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def PCA(X, d):
    # 将每个样本中心化，将特征值规范化
    X = X - np.mean(X, axis=0)
    
    # 计算协方差矩阵
    cov_matrix = np.cov(X, rowvar=False)
    
    # 对协方差矩阵做特征值分解
    eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
    
    # 获取最大的d个特征值所对应的特征向量的索引
    idx = np.argsort(eigenvalues)[::-1][:d]
    
    # 选择最大的d个特征值所对应的特征向量
    principal_eigenvectors = eigenvectors[:, idx]
    
    # 对原始数据进行投影
    X_pca = np.dot(X, principal_eigenvectors)
    
    return X_pca
```

```python
data = np.array([[1, 2.5], [2, 3], [3, 3.5], [4, 4.5], [5, 7]])
plt.scatter(data[:, 0], data[:, 1], c='b', marker='o', label='data')
plt.show()
```

```python
PCA_data = PCA(data, 1)
PCA_data
```

```python
plt.scatter(PCA_data, np.zeros(PCA_data.shape), c='r', marker='o', label='PCA_data')
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=1)
pca.fit(data)
projected_data= pca.transform(data)
projected_data
```

```python
plt.scatter(projected_data, np.zeros(projected_data.shape), c='r', marker='o', label='PCA_data')
plt.show()
```
